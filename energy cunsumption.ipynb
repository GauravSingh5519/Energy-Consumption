### **1. PLOTTING**

```python
import pandas as pd;
csvFile = pd.read_csv("C:\\Users\\Gaurav\\Desktop\\AIF\\python datasets\\long_data_.csv")
print(csvFile)
result = csvFile.groupby(['States', 'Dates'])['Usage'].sum()
print(result)
```

```python
# Creating a new column 'WinningStreak' to track the winning streak for each team
df['Maxusage'] = df['Dates'].ne(df['Usage'].shift()).cumsum()
```

```python
# Grouping the data by 'States' and 'max usages' and counting the occurrences
Max_usage = df.groupby(['Dates', 'Maxusage']).size().reset_index(name='Count')
```

```python
result = csvFile.groupby(['States', 'Regions'])['Usage'].sum()
print(result)
```

```python
# Plotting the hourly tweet activity
plt.figure(figsize=(10, 6))
result.plot(kind='line', color='darkblue')
plt.xlabel('Hour of the Day')
plt.ylabel('Tweet Count')
plt.title('Hourly Tweet Activity')
plt.xticks(range(24))
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
